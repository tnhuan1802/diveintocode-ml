```python
plt.title('Pie chart of the number of samples per species')
plt.pie(df['Species'].value_counts(), labels=data['target_names'])
plt.show()
```

While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

```python
fig, ax = plt.subplots(1,1, figsize=(9, 4))
bp1 = ax.boxplot([df.loc[:,'sepal_length'], df.loc[:,'sepal_width'], df.loc[:,'petal_length'], df.loc[:,'petal_width']], vert = True, patch_artist = True, labels = df.columns.values[:4])
colors = ['red', 'green', 'blue', 'pink']

for patch, color in zip(bp1['boxes'], colors):
    patch.set_facecolor(color)
```

```python
fig, (ax1, ax2) = plt.subplots(2,2, figsize=(10,8))
vp1 = ax1[0].violinplot(df['sepal_length'],vert = False)
vp2 = ax1[1].violinplot(df['sepal_width'],vert = False)
vp3 = ax2[0].violinplot(df['petal_length'],vert = False)
vp4 = ax2[1].violinplot(df['petal_width'],vert = False)
```

Problem 7:

```python
names = ['sepal_length', 'sepal_width','petal_length','petal_width']
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
pos = 0
for i in names:
  for j in names:
    if i == j: break
    sns.scatterplot(data=df, x=i, y=j, ax=axes[floor(pos/3), pos % 3], hue='Species')
    pos = pos + 1
```

```python
pp = sns.pairplot(data=df, hue="Species")
```

```python
corr_matrix = df.iloc[:,:-1].corr(method='pearson')
print(corr_matrix)
```

```python
hm = sns.heatmap(data=corr_matrix)
```
