```
import numpy as np
import matplotlib.pyplot as plt
```

Problem 1:

```
mean = [-3, 0]
cov  = [[1.0,0.8], [0.8, 1.0]]

data = np.random.multivariate_normal(mean, cov, 500)
# print(data)
np.random.seed(0)
```

Problem 2:

```
np.random.seed(0)
plt.scatter(data[:,0], data[:,1])
plt.show()
```

Problem 3:

```
plt.hist(data[:,0])
plt.xlabel("Frequency")
plt.xlim(-6, 1)
plt.show()

plt.hist(data[:,1])
plt.xlabel("Frequency")
plt.xlim(-5, 5)
plt.show()
```

Problem 4:

```
mean = [0, -3]
cov  = [[1.0,0.8], [0.8, 1.0]]
np.random.seed(0)
data2 = np.random.multivariate_normal(mean, cov, 500)
```

```
plt.scatter(data[:,0], data[:,1])
plt.scatter(data2[:,0], data2[:,1])
plt.legend(['0', '1'])
plt.show()
```

Problem 5:

```
new_data = np.concatenate((data, data2))
new_data
```

Problem 6:

```
data = np.concatenate((data, np.zeros((data[:, 1].size, 1))), axis=1)
data2 = np.concatenate((data2, np.ones((data2[:, 1].size, 1))), axis=1)
new_data = np.concatenate((data, data2))
new_data
```
