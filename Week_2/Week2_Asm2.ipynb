```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
```

```python
csv_path = "/mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
position = fuji[:, 0]
lattitude = fuji[:, 1]
longitude = fuji[:, 2]
elevation = fuji[:, 3]
```

Problem 2:

```python

x = position
y = elevation
def calculateSlope(currentPointNumber):
  if currentPointNumber == 0 or currentPointNumber == len(position): return 0
  yDiff = y[currentPointNumber] - y[currentPointNumber - 1]
  xDiff = x[currentPointNumber] - x[currentPointNumber - 1]

  return yDiff / xDiff

slope = calculateSlope(6)
print(slope)
```

Problem 3

```python
def calculateDest(currentPointNumber, alpha):
  try:
    val = currentPointNumber - alpha * calculateSlope(currentPointNumber)
    if val < 0: raise Exception("Illegal")
    return ceil(val)
  except Exception as e:
    print(x)
```

Problem 4

```python
def goDownTheMountain(initial, alpha):
  dest = calculateDest(initial, alpha)
  points = [initial]
  step, maxstep = 0, 100
  while step < maxstep:
    temp = calculateDest(dest, alpha)
    if temp == dest: break;
    else: 
      points.append(temp)
      dest = temp
    step = step + 1
  return points
```

Problem 5

```python
lst = goDownTheMountain(136, 0.2)
def drawGraph(go_down):
  plt.xlabel('position')
  plt.ylabel('elevation[m]')
  plt.plot(position, elevation)
  plt.plot(go_down, elevation[go_down], 'og')
  plt.show()

  plt.xlabel('iteration')
  plt.ylabel('altitude[m]')
  plt.plot(np.arange(0, len(go_down)), elevation[go_down])
  plt.show()

  slope_lst = [calculateSlope(i) for i in go_down]
  plt.xlabel('iteration')
  plt.ylabel('slope')
  plt.plot(np.arange(0, len(go_down)), slope_lst)
  plt.show()

drawGraph(lst)
```
