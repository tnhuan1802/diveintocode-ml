```python
import matplotlib.pyplot as plt
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")

THICKNESS = 0.00008
FOLDED_TIMES = 43
records = [THICKNESS]
folded_thickness = THICKNESS
for i in range(FOLDED_TIMES):
    folded_thickness *= 2
    records.append(folded_thickness)
plt.plot(records)
plt.show()
```
