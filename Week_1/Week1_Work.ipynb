```python
import time
import matplotlib.pyplot as plt


## Initialization
THICKNESS = 0.00008
FOLDED_TIMES = 800
records = [THICKNESS]
first_method_time = 0
second_method_time = 0
# Problem 1
print("Problem 1:")
start_1 = time.time()
folded_thickness = THICKNESS * (2 ** FOLDED_TIMES)
first_method_time = time.time() - start_1
print("Thickness: {} meters".format(folded_thickness))
# Problem 2
print("Problem 2:")
print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
# Problem 3
print("Problem 3:")
folded_thickness3 = THICKNESS
start_2 = time.time()
for i in range(FOLDED_TIMES):
    folded_thickness3 *= 2
    records.append(folded_thickness3)
second_method_time = time.time() - start_2
print("Thickness: {} meters".format(folded_thickness3))
# Convert meters to kilometers and display with two decimal places
print("Thickness: {: .2f} kilometers".format(folded_thickness3/1000))

# Problem_4
print("Problem 4:")
print("First method time : {}[s]".format(first_method_time))
print("Second method time : {}[s]".format(second_method_time))

# Problem_5
print("Problem 5:")
print(len(records))

# Problem_6 and 7
print("Problem 6 and 7:")
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(records)

plt.tick_params(direction='out', length=6, width=2, colors='r')
plt.show()
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(records,'or')
plt.show()
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(records, 'xb')
plt.show()
```
